```python

import numpy as np
import matplotlib.pyplot as plt
#######################################################
# This program creates a plot of the trajectory of a 
# grenade being thrown from a cliff 3000m above with
# an initial velocity of v0 = (4,4)m/s. At time t = 4,
# the grenade detonates, creating a second fragment
# that shoots off at a velocity v + dv, where dv is 
# given as dv = (1, 3)m/s. This program does include
# AIR RESISTANCE
#
# Run the program to show a graph of the trajectories
# **The graph created by this program shows x and y
#   values at any point that you put the cursor over.
#   This doesn't work in the Jupyter notebook environment
#   but if the program is run elsewhere, a separate window
#   will appear with this functionality**
#######################################################

# Initial Values (Some can be manipulated because
# they are not explicitly known in the problem, they
# will be denoted with a comment.)

vx = 4
vy = 4
g = 9.81
hi = 3000
c = 0.1  # the quadratic drag constant
m = 1  # the mass of the grenade being thrown

# Next the time will be declared with a Numpy range function
# that gives several values from a starting to ending point.
# The third argument given will help to smooth the curve by
# giving more points in the given range.

t = np.arange(0.0, 700.0, 0.1)

# To help clean up some of the code, some variables to input
# into the position functions will be declared next.

V = np.sqrt((m * g) / c)
H = ((hi * g) / (np.power(V, 2)))

# Next is the input of the position functions of a projectile
# subject to a quadratic drag force

xt = (np.log(1 + ((vx * t * g) / np.power(V, 2)))) * ((np.power(V, 2)) / g)

yt = (((-1 * t * g) / (vy * 2)) - ((np.power(t, 2) * np.power(g, 2)) / (4 * np.power(V, 2))) +
      (((vy / vx) + (1 / (2 * (vx * vx) / np.power(V, 2)))) * np.log(1 + ((vx * t * g) / (np.power(V, 2))))) + H) *(np.power(V, 2) / g)


# These equations can then be used to find their location when t = 4,
# so that the same equations (will need to be manipulated for vx and vy
# because they are no longer equal to 4) can then be applied to the
# second fragment of the grenade.

t_detonation = 4

x_detonation = (np.log(1 + ((4 * t_detonation * g) / np.power(V, 2)))) * ((np.power(V, 2)) / g)

y_detonation = (((-1 * t_detonation * g) / 8) - ((np.power(t_detonation, 2) * np.power(g, 2)) / (4 * np.power(V, 2))) +
                ((1 + (np.power(V, 2) / 32)) * np.log(1 + ((4 * t_detonation * g)
                 / (np.power(V, 2))))) + H) * (np.power(V, 2) / g)

# print(x_detonation)
# print(y_detonation)

# Next is the position functions for the second fragment sent away
# after detonation. A new time range needs to be declared for this
# new set of functions, as well a new "H" declaration.

vx_d = vx + 1
vy_d = vy + 3

t_d = np.arange(4.0, 700.0, 0.1)
h_d = y_detonation
H_d = ((h_d * g) / (np.power(V, 2)))

xt_d = (np.log(1 + ((vx_d * t_d * g) / np.power(V, 2)))) * ((np.power(V, 2)) / g)

yt_d = (((-1 * t_d * g) / (vy_d * 2)) - ((np.power(t_d, 2) * np.power(g, 2)) / (4 * np.power(V, 2))) +
        (((vy_d / vx_d) + (1 / (2 * (vx_d * vx_d) / np.power(V, 2)))) * np.log(1 + ((vx_d * t_d * g) / (np.power(V, 2)))))
        + H_d) * (np.power(V, 2) / g)


# Next is the plot function imported by the matplotlib.pyplot
# library. Arguments will be xt and yt for first fragment and
# xt_d and yt_d for detonated fragment.

fig = plt.figure()
# plt.xkcd()
plt.plot(xt, yt, label='first fragment')
plt.plot(xt_d, yt_d, label='second fragment')
plt.grid(color='b', linestyle='-', linewidth=0.5)
plt.xlim(0, 50)
plt.ylim(0, 3200)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grenade Thrown And Detonating With Drag Force')
plt.show()
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
#######################################################
# This program creates a plot of the trajectory of a 
# grenade being thrown from a cliff 3000m above with
# an initial velocity of v0 = (4,4)m/s. At time t = 4,
# the grenade detonates, creating a second fragment
# that shoots off at a velocity v + dv, where dv is 
# given as dv = (1, 3)m/s. This program does NOT include
# AIR RESISTANCE
# 
# Run the program to show the graph of the trajectories
#######################################################

# Initial Values (Some can be manipulated because
# they are not explicitly known in the problem, they
# will be denoted with a comment.)

x0 = 0
y0 = 3000
vx1 = 4
vy1 = 4
vx2 = 1
vy2 = 3
g = -9.81
m = 1  # the mass of the grenade being thrown

# Next the time will be declared with a Numpy range function
# that gives several values from a starting to ending point.
# The third argument given will help to smooth the curve by
# giving more points in the given range.

t = np.arange(0.0, 700.0, 0.1)

# Next is the input of the position functions of a projectile
# subject to NO quadratic drag force

xt = x0 + (vx1 * t)

yt = y0 + (vy1 * t) + ((1 / 2) * g * np.power(t, 2))

# These equations can then be used to find their location when t = 4.
# These positions can then be used for the second fragments of the
# grenade.

t_detonation = 4

xt_detonation = x0 + (vx1 * t_detonation)

yt_detonation = y0 + (vy1 * t_detonation) + ((1 / 2) * g * np.power(t_detonation, 2))

# Next is the position functions for the second fragment sent away
# after detonation. A new time range needs to be declared for this
# new set of functions, as well a new y and x declaration.

vx0_d = vx1 + 1
vy0_d = vy1 + 3

t_d = np.arange(4.0, 7000.0, 0.1)
x0_d = x_detonation
y0_d = y_detonation


xt_d = x0_d + (vx0_d * t_d)

yt_d = y0_d + (vy0_d * t_d) + ((1 / 2) * g * np.power(t_d, 2))

# Next is the plot function imported by the matplotlib.pyplot
# library. Arguments will be xt and yt for first fragment and
# xt_d and yt_d for detonated fragment.

fig = plt.figure()
# plt.xkcd()
plt.plot(xt, yt, label='first fragment')
plt.plot(xt_d, yt_d, label='second fragment')
plt.grid(color='b', linestyle='-', linewidth=0.5)
plt.xlim(0, 175)
plt.ylim(0, 4000)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grenade Thrown And Detonating Without Drag Force')
plt.show()
```
